# Horizontally Stacked Bar Graph

```python
month = ['Jan','Feb','Mar','Apr','May']
HYDsales = [50,60,55,75,65]
BANGsales = [55,50,75,60,80]
index = np.arange(1,6)
width = 0.30
plt.bar(index, HYDsales, width, color='green', label= 'Hyd Sales')
plt.bar(index+width, BANGsales,width, color = 'red', label= 'BangSales')
plt.title('Horizontally stacked Bar Graph')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(index+width/2, month)
plt.legend(loc='best')
plt.show()
```

```python
month = ['Jan','Feb','Mar','Apr','May']
HYDsales = [50,60,55,75,65]
BANGsales = [55,50,75,60,80]
index = np.arange(1,6)
width = 0.30
plt.barh(index, HYDsales, width, color='green', label= 'Hyd Sales')
plt.barh(index+width, BANGsales,width, color = 'red', label= 'BangSales')
plt.title('Horizontally stacked Bar Graph')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(index+width/2, month)
plt.legend(loc='best')
plt.show()
```

```python
month = ['Jan','Feb','Mar','Apr','May']
HYDsales = [50,60,55,75,65]
BANGsales = [55,50,75,60,80]
index = np.arange(1,6)
width = 0.30
plt.bar(index, HYDsales, width, color='green', label= 'Hyd Sales')
plt.bar(index+width, BANGsales,width, color = 'red', label= 'Bang Sales', bottom= HYDsales)
plt.title('Horizontally stacked Bar Graph')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(index+width/2, month)
plt.legend(loc='best')
plt.show()
```

# multiple plot in one graph
 

subplot (row, column, position)

```python
plt.subplot(1,2,1)
plt.plot([1,2,3,4],[10,20,30,40], 'go')
plt.title('1st plot')

x= np.arange(1,5)
y= x**2

plt.subplot(1,2,2)
plt.plot(x,y)
plt.title('2nd plot')
plt.suptitle('Sales Plots')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot([1,2,3,4],[10,20,30,40], 'go')
plt.title('1st plot')

x= np.arange(1,5)
y= x**2

plt.subplot(2,1,2)
plt.plot(x,y)
plt.title('2nd plot')
plt.suptitle('Sales Plots')
plt.show()
```
